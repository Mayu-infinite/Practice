```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"

column_names = ["sepal length", "sepal width", "petal length", "petal width", "species"]
iris_df = pd.read_csv(url, header=None, names=column_names)
species_mapping = {species: idx for idx, species in enumerate(iris_df['species'].unique())}
iris_df['species numeric'] = iris_df['species'].map(species_mapping)
x = iris_df['sepal length']
y = iris_df['petal length']
sizes = iris_df['sepal width'] *20 #for_visibility
colors = iris_df['species numeric']

plt.figure(figsize=(10, 6))
scatter = plt.scatter(x, y, s=sizes, c=colors, cmap='Blues', marker='o', edgecolor='black', linewidth=1, alpha=0.95)
cbar = plt.colorbar(scatter)
cbar.set_label('Species')
cbar.set_ticks(np.arange(len(species_mapping)))
cbar.set_ticklabels(list(species_mapping.keys()))

plt.title('Scatter Plot of Sepal Length vs Petal Length')
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

url = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"

column_names = ["sepal length", "sepal width", "petal length", "petal width", "species"]

iris_df = pd.read_csv(url, header=None, names=column_names)
species_names = iris_df['species'].unique()
features = ['sepal length', 'sepal width', 'petal length', 'petal width']
colors = ['#FF9999', '#66B2FF', '#99FF99']

for feature in features:
    plt.figure(figsize=(10, 6))

    data_to_plot = [iris_df[iris_df['species'] == species][feature] for species in species_names]
    plt.boxplot(data_to_plot, labels=species_names)

    plt.title(f'Box Plot of {feature} by Species')
    plt.xlabel('Species')
    plt.ylabel(feature)

    plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

url = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"

# columns
column_names = ["sepal length", "sepal width", "petal length", "petal width", "species"]

iris_df = pd.read_csv(url, header=None, names=column_names)

selected_features = iris_df[["sepal length", "petal length"]]
correlation_matrix = selected_features.corr()
plt.imshow(correlation_matrix, cmap='PuBu', vmin=-1, vmax=1)
plt.colorbar()
plt.xticks(ticks=np.arange(len(selected_features.columns)), labels=selected_features.columns)
plt.yticks(ticks=np.arange(len(selected_features.columns)), labels=selected_features.columns)

# Adding the correlation values on the heatmap
for i in range(len(selected_features.columns)):
    for j in range(len(selected_features.columns)):
        plt.text(j, i, f"{correlation_matrix.iloc[i, j]:.2f}", ha='center', va='center', color='black')

plt.title("Correlation Heatmap")
plt.show()
```
