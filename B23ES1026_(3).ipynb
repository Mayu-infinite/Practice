1. Applying PCA from scratch:
Implement Principal Component Analysis from scratch. Make sure that it should:
○ Be a class that stores the eigenvectors and eigenvalues obtained.
○ Should store explained variance for the components.
○ Should derive the covariance matrix from scratch.
○ Should compute Eigenvectors, Eigenvalues and Principal components.
(You may use scikit-learn to find the eigenvectors but others are to be
found from scratch.)
○ Should have a fit function to fit the PCA model.
○ Should have a transform function to project the input data to principal
components. (takes “n = number of principal components” as input)

```python
#  initiating with defining a class
class PCA:

    def __init__(self, n_components):
        self.n_components = n_components
        self.components = None
        self.mean = None
        self.variance = None

    def fit(self, x): # the fit function to fit PCA model
        # mean centering that is difference between the value and the mean
        self.mean = np.mean(x, axis = 0)
        x = x - self.mean

        # calculating covariance from scratch by taking summation of dot products of arrays and finally dividing it by n-1
        cov = np.dot(x.T,x) / (x.shape[0] - 1)

        # eigenvectors and eigenvalues are getting stored in the variables eigenvec and eigenval
        eigvec, eigval = np.linalg.eig(cov)


        # sort the indixes of eigenvalues
        idx = np.argsort(eigval)[::-1]
        # re assigning the values to eigenvectors and eigenvalues according to the sorted indices basically to have values with greater variance
        eigval = eigval[idx]
        eigvec =eigvec[idx]

        self.components = eigvec[:self.n_components]
        self.variance = eigval[:self.n_components]/np.sum(eigval)

    def transform(self, x):
        x = x - self.mean
        return np.dot(x,self.components.T)
```

2. Use the above-made PCA to reduce the data up to a chosen

```python
# Executing the PCA
pca = PCA(2)
pca.fit(d1)
new_data = pca.transform(d1)
print(d1.shape)
print(new_data.shape)
```

```python
# Plot a scatter plot to show the direction of the eigenvectors along with the data points
plt.scatter(new_data[:, 0], new_data[:, 1], alpha=0.5)
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA Scatter Plot')
plt.show()
```

3. Plot a bar graph to show the change in variance as you increase the no. of

```python
# trying different values of n_components and printing a bar graph to show the variance explained by each component
var = []
for n in range(3,10):
    pca = PCA(n)
    pca.fit(d1)
    new_data = pca.transform(d1)
    var.append(np.sum(pca.variance))
plt.bar(range(3,10), var)
plt.xlabel('Number of Components')
plt.ylabel('Variance Explained')
plt.title('Variance Explained by each Component')
plt.show()
```

## Question 3.
1. Take a classifier and perform comparative analysis on train and test results for
different values of “n”.


```python
# segrigating in train and test data and changing the y to categorical
x_train = d1[:, :-1]
y_train = d1[:, -1]
x_test = d2[:, :-1]
y_test = d2[:, -1]
y_train = pd.cut(y_train, bins=3, labels=[0, 1, 2])
y_test = pd.cut(y_test, bins=3, labels=[0, 1, 2])
y_train = np.array(y_train)
y_test = np.array(y_test)
```

```python
# applying the classifier fro different values of n
from sklearn.metrics import accuracy_score
train_accu = []
test_accu = []
for n in range(1,10):
    pca = PCA(n)
    pca.fit(x_train)
    x_train_pca = pca.transform(x_train)
    x_test_pca = pca.transform(x_test)

    # applying the classifier
    from sklearn.linear_model import LogisticRegression
    lr = LogisticRegression()
    lr.fit(x_train_pca, y_train)

    #  predicting the values
    y_train_pred = lr.predict(x_train_pca)
    y_test_pred = lr.predict(x_test_pca)

    # calculating the accuracy
    train_accu.append(accuracy_score(y_train, y_train_pred))
    test_accu.append(accuracy_score(y_test, y_test_pred))
```

```python
# printing the accuracy
print('Accuracy score for trainig data', train_accu)
print('Accuracy score for testing data', test_accu)
```
